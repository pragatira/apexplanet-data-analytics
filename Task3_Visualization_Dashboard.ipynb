```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.dpi"] = 100

df = pd.read_csv("../data/ecommerce_sales_cleaned.csv")
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
print("✅ Data loaded!", df.shape)
```

## Day 14-15: Matplotlib Charts

```python
# Line Chart - Monthly Sales Trend
fig, ax = plt.subplots(figsize=(10, 5))
monthly = df.groupby(df["Order_Date"].dt.to_period("M"))["Net_Sales"].sum()
monthly.index = monthly.index.astype(str)
ax.plot(monthly.index, monthly.values, marker="o", color="#4C72B0", linewidth=2)
ax.set_title("Monthly Net Sales Trend", fontsize=14, fontweight="bold")
ax.set_ylabel("Net Sales (₹)")
ax.set_xlabel("Month")
plt.xticks(rotation=60, ha="right")
plt.tight_layout()
plt.savefig("../reports/chart_line_monthly.png", dpi=150)
plt.show()
print("✅ Line chart saved!")
```

```python
# Bar Chart - Sales by Category
fig, ax = plt.subplots(figsize=(9, 5))
cat_sales = df.groupby("Category")["Net_Sales"].sum().sort_values(ascending=False)
colors = ["#4C72B0","#DD8452","#55A868","#C44E52","#8172B2","#64B5CD"]
cat_sales.plot(kind="bar", ax=ax, color=colors, edgecolor="white")
ax.set_title("Total Net Sales by Category", fontsize=14, fontweight="bold")
ax.set_ylabel("Net Sales (₹)")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.savefig("../reports/chart_bar_category.png", dpi=150)
plt.show()
print("✅ Bar chart saved!")
```

```python
# Scatter Plot - Unit Price vs Net Sales
fig, ax = plt.subplots(figsize=(8, 5.5))
sample = df.sample(800, random_state=1)
ax.scatter(sample["Unit_Price"], sample["Net_Sales"], alpha=0.4, c="#55A868", edgecolors="none")
ax.set_xlabel("Unit Price (₹)")
ax.set_ylabel("Net Sales (₹)")
ax.set_title("Unit Price vs Net Sales", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/chart_scatter.png", dpi=150)
plt.show()
print("✅ Scatter chart saved!")
```

```python
# Histogram - Order Quantity Distribution
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df["Quantity"], bins=range(1, 8), color="#8172B2", edgecolor="white", align="left")
ax.set_title("Order Quantity Distribution", fontsize=14, fontweight="bold")
ax.set_xlabel("Quantity")
ax.set_ylabel("Number of Orders")
plt.tight_layout()
plt.savefig("../reports/chart_hist_quantity.png", dpi=150)
plt.show()
print("✅ Histogram saved!")
```

## Seaborn Advanced Charts

```python
# Heatmap - Region x Category
fig, ax = plt.subplots(figsize=(9, 6))
pivot = df.pivot_table(values="Net_Sales", index="Region", columns="Category", aggfunc="sum")
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu", ax=ax, linewidths=0.5)
ax.set_title("Sales Heatmap: Region × Category", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/chart_heatmap.png", dpi=150)
plt.show()
print("✅ Heatmap saved!")
```

```python
# Pairplot
pp_sample = df[["Quantity", "Unit_Price", "Discount_Percent", "Net_Sales"]].sample(500, random_state=1)
g = sns.pairplot(pp_sample, diag_kind="kde", plot_kws={"alpha": 0.4, "s": 18})
g.fig.suptitle("Pairplot of Key Numeric Features", y=1.02, fontsize=13)
plt.savefig("../reports/chart_pairplot.png", dpi=100, bbox_inches="tight")
plt.show()
print("✅ Pairplot saved!")
```

```python
# Boxen Plot
fig, ax = plt.subplots(figsize=(9, 5.5))
sns.boxenplot(data=df, x="Customer_Segment", y="Net_Sales", ax=ax, palette="Set2")
ax.set_title("Net Sales by Customer Segment", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/chart_boxen.png", dpi=150)
plt.show()
print("✅ Boxen plot saved!")
```
